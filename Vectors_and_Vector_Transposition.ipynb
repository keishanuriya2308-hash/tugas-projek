2. akukan transpose pada vector x dan simpan pada variabel x_t

```python
# lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
x_t
```

```python
# tampilkan ukuran dari vector x_t
x_t.shape
```

```python
# lakukan transpose pada vector y dan simpan pada variabel y_t
y = np.array([[1],
              [0],
              [0]])
y_t = y.T
y_t
```

```python
# tampilkan ukuran dari vector y_t
y_t.shape
```

3. lakukan soal nomor 2 menggunakan pytorch dan tensorflow

```python
import torch as pt
```

```python
x = pt.tensor([[1],
               [0],
               [0]])
x
```

```python
x_t = x.T
x_t
```

```python
x_t.shape
```

```python
y = pt.tensor([[1],
               [0],
               [0]])
y
```

```python
y_t = y.T
y_t
```

```python
import tensorflow as tf
```

```python
x = tf.constant([[1],
                 [0],
                 [0]])
x
```

```python
x_t = tf.transpose(x)
x_t
```

```python
y = tf.constant([[1],
                 [0],
                 [0]])
y
```

```python
y_t = tf.transpose(y)
y_t
```
